```python
# Task2
orders=[1200,2500,800,1750,3000]
print(f"Order Amount \t Discount Amount\t Sub Total \t Final amount")
print("----------------------------------------------------------------------")
discount_count=0
for order_amount in orders:
  if order_amount >=2000:
    discount="15%"
    discount_count+=1
    order_amount_updated=order_amount-0.15
  elif order_amount >= 1500 and order_amount<2000:
    discount="10%"
    discount_count+=1
    order_amount_updated=order_amount-0.10
  elif order_amount >= 1000 and order_amount<1500:
    discount="7%"
    discount_count+=1
    order_amount_updated=order_amount-0.07
  Final_amount=order_amount+0.05
  print(f"{order_amount}\t\t\t {discount}\t\t {order_amount_updated}       \t {Final_amount}")


print("Number of orders with discount greater that 0%",discount_count)
```
