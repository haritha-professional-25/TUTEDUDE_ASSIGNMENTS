```python
# Task3
price_dict={"Pen":10,"Book":30,"Pencil":5,"Paper":100,"Sharpner":10,"Scale":15}
print("Product dictionary",price_dict)
price_dict["Sketch"]=50
price_dict.update({"Scale":20})
price_dict.pop("paper".capitalize())
print("Updated dictionary",price_dict)
print(f"Maximum price product {max(price_dict,key=price_dict.get)}")
print(f"Minimum price product {min(price_dict,key=price_dict.get)}")
```
