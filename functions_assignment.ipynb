```python
#TASK 1
def apply_discount(price,discount_percent=5):
  return price-price*(discount_percent/100)

print(apply_discount(1000,10))
print(apply_discount(500))

```

```python
#TASK 2
def factorial(n):
  if n<0:
    return "Error"
  elif n==0 or n==1:
    return 1
  else:
    return n*factorial(n-1)
print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

```python
# TASK 6
prices=[100,250,400,1200,50,2000,850]

def process_prices(price):
  return list(map(lambda price:price+price*0.10,prices)),list(filter(lambda amount : amount>300,price))

discounted_prices,filtered_prices=process_prices([100,500,900,50,750])
print(f"Discounted price {discounted_prices},Filtered price {filtered_prices}")
```

```python
#TASK 7
def add_price(prices_list,price):
  prices_list.append(price)
  return prices_list
def get_average_price(prices_list):
  return sum(prices_list)/len(prices_list)
def get_max_price(prices_list):
  return max(prices_list)

prices=[100,250,400,1200,50,2000,850]
while True:
  print(f" 1. ADD PRICE \n 2.SHOW AVERAGE PRIC \n 3.SHOW HIGHEST PRICE \n q.QUIT")
  choice=input("Enter your choice")
  if choice== "1":
    print("List after updation",add_price(prices,700))
  elif choice =="2":
    print("Average price",get_average_price(prices))
  elif choice =="3":
    print("Maximum price",get_max_price(prices))
  elif choice =="q":
    print("THANK YOU")
    break
  else:
    print("Invalid entry")
    continue


```
